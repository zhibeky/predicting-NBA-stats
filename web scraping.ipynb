```python
dfs = []
for year in years:
    with open('mvp/{}.html'.format(year), 'r', encoding='utf-8') as f:
        page = f.read()
    soup = bs(page, "html.parser")
    soup.find('tr', class_='over_header').decompose()
    mvp_table = soup.find(id='mvp')
    mvp = pd.read_html(str(mvp_table), flavor='bs4')[0]
    mvp["Year"] = year
    
    dfs.append(mvp)
```

```python
df = []
for year in years:
    
    with open("player/{}.html".format(year), encoding='utf-8') as f:
        page = f.read()

    soup = bs(page, "html.parser")
    soup.find('tr', class_='thead').decompose()
    player_table = soup.find(id='per_game_stats')
    player = pd.read_html(str(player_table), flavor='bs4')[0]
    player["Year"] = year
    df.append(player)
```

```python
players = pd.concat(df)
```

```python
players
```

```python
players.to_csv('players.csv', index=False)
```

```python
team_stats_url = 'https://www.basketball-reference.com/leagues/NBA_{}_standings.html'
```

```python
for year in years:
    url = team_stats_url.format(year)

    data = requests.get(url)

    with open('team/{}.html'.format(year), 'w+', encoding='utf-8') as f:
        f.write(data.text)
```

```python
dfs = []
for year in years:
    with open("team/{}.html".format(year), encoding='utf-8') as f:
        page = f.read()

    soup = bs(page, "html.parser")
    soup.find('tr', class_='thead').decompose()
    team_table = soup.find(id='divs_standings_E')
    team = pd.read_html(str(team_table), flavor='bs4')[0]
    team["Year"] = year
    team["Team"] = team["Eastern Conference"]
    del team["Eastern Conference"]
    dfs.append(team)

    soup = bs(page, "html.parser")
    soup.find('tr', class_='thead').decompose()
    team_table = soup.find(id='divs_standings_W')
    team = pd.read_html(str(team_table), flavor='bs4')[0]
    team["Year"] = year
    team["Team"] = team["Western Conference"]
    del team["Western Conference"]
    dfs.append(team)
```

```python
teams = pd.concat(dfs)
```

```python
teams
```

```python
teams.to_csv('teams.csv', index=False)
```
